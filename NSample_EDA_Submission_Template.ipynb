### Variables Description

The dataset contains 135 columns providing detailed information about global terrorism events. Here's a simplified overview of some key variables:

*   **Event Identifiers (`eventid`)**: Unique codes for each attack.
*   **Temporal Information (`iyear`, `imonth`, `iday`)**: When the attack occurred (year, month, day). Note that some months/days are recorded as 0, likely indicating unknown values.
*   **Geographical Information (`country_txt`, `region_txt`, `latitude`, `longitude`)**: Where the attack took place. There are some missing geographical coordinates.
*   **Attack Details (`attacktype1_txt`, `targtype1_txt`, `weaptype1_txt`)**: The method of attack, target type, and weapon used. These are categorical descriptions.
*   **Perpetrator Information (`gname`)**: The name of the group responsible for the attack.
*   **Casualties (`nkill`, `nwound`)**: Number of people killed and wounded. Many values are missing.
*   **Hostage/Ransom Details (`nhostkid`, `ransomamt`, `ransompaid`)**: Information related to hostage-taking and ransom, with many missing or placeholder values (e.g., -9, -99).
*   **International Incident Flags (`INT_LOG`, `INT_IDEO`, `INT_MISC`, `INT_ANY`)**: Indicators for the international nature of the incident, also containing placeholder values.

### Check Unique Values for each variable.

````python
# Check Unique Values for each variable.
for col in df.columns:
    unique_vals = df[col].nunique()
    print(f"Column '{col}': {unique_vals} unique values")
    if unique_vals < 50 and unique_vals > 0:
        print(f"    Unique values: {df[col].unique()}")
    elif unique_vals == 0:
        print("    (All values are missing/null)")
````

## 3. ***Data Wrangling***

### Data Wrangling Code

````python
df.rename(columns={'iyear':'Year','imonth':'Month','iday':'Day','country_txt':'Country','provstate':'State',
                   'region_txt':'Region','attacktype1_txt':'AttackType','target1':'Target','nkill':'Killed','nwound':'Wounded',
                   'summary':'Summary','gname':'Group','targtype1_txt':'Target_Type','weaptype1_txt':'Weapon_Type'},
                  inplace=True)
df = df[['Year','Month','Day','Country','State','Region','city','latitude','longitude','AttackType','Killed',
         'Wounded','Target','Summary','Group','Target_Type','Weapon_Type']]

missing_percentage = df.isnull().sum() / len(df) * 100
drop_columns = missing_percentage[missing_percentage > 50].index
df.drop(columns=drop_columns, inplace=True)

print(f"Dropped {len(drop_columns)} columns with more than 50% missing values: {list(drop_columns)}")
print("Remaining columns:", df.columns.tolist())
print("\nUpdated DataFrame info:")
df.info()
````

### What all manipulations have you done and insights you found?

The data wrangling steps performed were:

 Several columns were renamed to more intuitive names. This significantly enhances the readability and interpretability of the DataFrame.
 columns having more significance then other were displayed first.

Insights from these manipulations:
The dataset is now more streamlined and focused, containing only the columns deemed most relevant for the analysis, with clear and understandable names.The absence of columns with more than 50% missing values among the selected features suggests that these columns retain sufficient data for meaningful analysis, although individual rows within these columns may still have missing values

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('/content/drive/MyDrive/Global Terrorism Data.csv', encoding='latin1')

df.rename(columns={'iyear':'Year','imonth':'Month','iday':'Day','country_txt':'Country','provstate':'State',
                   'region_txt':'Region','attacktype1_txt':'AttackType','target1':'Target','nkill':'Killed','nwound':'Wounded',
                   'summary':'Summary','gname':'Group','targtype1_txt':'Target_Type','weaptype1_txt':'Weapon_Type'},
                  inplace=True)

df = df[['Year','Month','Day','Country','State','Region','city','latitude','longitude','AttackType','Killed',
         'Wounded','Target','Summary','Group','Target_Type','Weapon_Type']]

df['Killed'] = df['Killed'].fillna(0)
df['Wounded'] = df['Wounded'].fillna(0)


# Chart - 15 visualization code

pair_plot_data = df[['Year', 'Killed', 'Wounded', 'latitude', 'longitude']]

pair_plot_data = pair_plot_data.dropna()

sample_size = 2000
if len(pair_plot_data) > sample_size:
    pair_plot_data_sampled = pair_plot_data.sample(n=sample_size, random_state=42)
else:
    pair_plot_data_sampled = pair_plot_data

plt.figure(figsize=(15, 15))
sns.pairplot(pair_plot_data_sampled)
plt.suptitle('Pair Plot of Numerical Variables (Sampled Data)', y=1.02)
plt.show()
````

##### 1. Why did you pick the specific chart?

A pair plot (or scatterplot matrix) is an excellent choice for visualizing the pairwise relationships and distributions of multiple numerical variables simultaneously. For each pair of variables, it creates a scatter plot, and for single variables, it displays their distribution (e.g., a histogram or KDE plot). This allows for a quick, high-level overview of potential correlations, patterns, and outliers across several dimensions.

##### 2. What is/are the insight(s) found from the chart?

This pair plot will allow us to observe the distribution of each variable (Year, Killed, Wounded, latitude, longitude) individually, and how each pair of variables scatters against each other. For instance, we can visually inspect the relationship between 'Killed' and 'Wounded' again, see the distribution of attacks over 'Year', and identify if there are any clusters or specific patterns in geographical coordinates ('latitude', 'longitude') related to casualties.
