**NUMPY**

```python
import numpy as np
import pandas as pd
df=pd.read_csv("/content/drive/MyDrive/International_T20_Data.csv")
```

```python
city_counts = df['info.city'].value_counts()

print(city_counts)
```

```python
arr = np.array(city_counts)

print("NumPy Array:")
print(arr)
```

```python
print("Mean matches:", np.mean(arr))
print("Maximum matches:", np.max(arr))
print("Minimum matches:", np.min(arr))
print("Total matches:", np.sum(arr))
```

```python
print("Standard Deviation:", np.std(arr))
```

```python
cities = city_counts.index

print("Cities:", cities)
print("Matches:", arr)
```
