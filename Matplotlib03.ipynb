```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("/content/drive/MyDrive/International_T20_Data.csv")
```

```python
city_counts = df['info.city'].value_counts().head(10)

print(city_counts)
```

```python
plt.bar(city_counts.index, city_counts.values)

plt.title("Top 10 Cities Hosting Matches")
plt.xlabel("City")
plt.ylabel("Number of Matches")

plt.xticks(rotation=45)  # rotate labels for clarity

plt.show()
```

```python
plt.barh(city_counts.index, city_counts.values)

plt.title("Top 10 Cities Hosting Matches")
plt.xlabel("Number of Matches")
plt.ylabel("City")

plt.show()
```

```python
plt.pie(city_counts.values, labels=city_counts.index, autopct='%1.1f%%')

plt.title("Match Distribution by City")

plt.show()
```
